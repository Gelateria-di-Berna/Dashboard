```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../docs/dataframes/dataframe.csv")
df["date"] = pd.to_datetime(df["date"])
```

```python
df
```

```python
df[df["location"] == "Brupbacherplatz"]
```

```python
df.info()
```

```python
df_daily_profit = pd.DataFrame(df.groupby([df["location"], df["date"].dt.date])["price"].sum())
```

```python
df_daily_profit = df_daily_profit.reset_index()
```

```python
for location in df_daily_profit['location'].unique():
    data = df_daily_profit[df_daily_profit['location'] == location]
    plt.plot(data['date'], data['price'], label=location)

plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df_hourly_profit = pd.DataFrame(df.groupby([df["location"], df["date"].dt.hour])["price"].mean())
```

```python
df_hourly_profit = df_hourly_profit.reset_index()
df_hourly_profit[df_hourly_profit["location"] == "Brupbacherplatz"]
```

```python
plt.bar(
    df_hourly_profit[df_hourly_profit["location"] == "Brupbacherplatz"]["date"],
    df_hourly_profit[df_hourly_profit["location"] == "Brupbacherplatz"]["price"]
)
plt.show()
```
