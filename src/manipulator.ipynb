```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../docs/dataframes/dataframe.csv")
df["date"] = pd.to_datetime(df["date"])
```

```python
df
```

```python
df.info()
```

```python
df_daily_profit = df.groupby(df["date"].dt.date)["price"].sum()
```

```python
plt.plot(
    df_daily_profit
)
plt.show()
```

```python
df_hourly_profit = df.groupby(df["date"].dt.hour)["price"].mean()
```

```python
plt.plot(df_hourly_profit)
plt.show()
```

```python
plt.bar(
    df_hourly_profit.index,
    df_hourly_profit
)
plt.show()
```
