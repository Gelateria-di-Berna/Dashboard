```python
import pandas as pd
import matplotlib.pyplot as plt
import src.data_loaders.hello_tess_loader
import datetime as dt
```

```python
df = src.data_loaders.hello_tess_loader.get_hello_tess_df()
```

```python
def filter_date_range(df: pd.DataFrame, start_date: dt.datetime, end_date: dt.datetime) -> pd.DataFrame:
    return df[(df["date"].dt.date >= start_date.date()) & (df["date"].dt.date <= end_date.date())]
```

```python
start = dt.datetime.strptime("01.04.2020", "%d.%m.%Y")
end = dt.datetime.strptime("10.04.2020", "%d.%m.%Y")
df_timerange = filter_date_range(df, start, end)
# Getting data for each location
for location in ["Brupbacherplatz", "Test"]:
    #Â Filter data for location
    df_location: pd.DataFrame = df_timerange[df_timerange["location"] == location]
    # Group
    df_hello_tess_hourly = pd.DataFrame(df_location.groupby(df_location["date"].dt.hour)["price"].mean().round(2))
    df_hello_tess_hourly.index.set_names("Stunde", inplace=True)
    df_hello_tess_hourly.rename(columns={"price":"Durchschnittlicher Umsatz"}, inplace=True)
    plt.bar(
        df_hello_tess_hourly.index.to_list(),
        df_hello_tess_hourly["Durchschnittlicher Umsatz"]
    )
plt.show()
```

```python
df_daily_profit = pd.DataFrame(df.groupby([df["location"], df["date"].dt.date])["price"].sum())
```

```python
df_daily_profit = df_daily_profit.reset_index()
```

```python
for location in df_daily_profit['location'].unique():
    data = df_daily_profit[df_daily_profit['location'] == location]
    plt.plot(data['date'], data['price'], label=location)

plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df_filtered = filter_date_range(df, start, end)

df_hourly_profit = pd.DataFrame(df_filtered.groupby([df_filtered["location"], df_filtered["date"].dt.hour])["price"].mean().round(2))
df_hourly_profit.index.set_names(["Standort", "Stunde"], inplace=True)
df_hourly_profit.rename(columns={"price":"Durchschnittlicher Umsatz"}, inplace=True)
df_hourly_profit
```

```python
plt.bar(
    df_hourly_profit[df_hourly_profit["Standort"] == "Brupbacherplatz"]["Stunde"],
    df_hourly_profit[df_hourly_profit["Standort"] == "Brupbacherplatz"]["Durchschnittlicher Umsatz"]
)
plt.show()
```
